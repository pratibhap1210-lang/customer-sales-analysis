```python
import pandas as pd

df = pd.read_csv("cleaned_superstore.csv")
df.head()
```

```python
import sqlite3

conn = sqlite3.connect("superstore.db")

df.to_sql("sales", conn, if_exists="replace", index=False)
```

```python
pd.read_sql_query("""
SELECT
    customer_name,
    ROUND(SUM(total_sales), 2) AS total_revenue
FROM sales
GROUP BY customer_name
ORDER BY total_revenue DESC
LIMIT 10;
""", conn)
```

```python
pd.read_sql_query("""
SELECT
    customer_name,
    COUNT(DISTINCT order_id) AS order_count
FROM sales
GROUP BY customer_name
ORDER BY order_count DESC
LIMIT 10;
""", conn)
```

```python
pd.read_sql_query("""
SELECT
    customer_name,
    COUNT(DISTINCT order_id) AS order_count
FROM sales
GROUP BY customer_name
HAVING COUNT(DISTINCT order_id) > 1
ORDER BY order_count DESC;
""", conn)
```

```python
pd.read_sql_query("""
SELECT
    customer_name,
    ROUND(SUM(profit), 2) AS total_profit
FROM sales
GROUP BY customer_name
ORDER BY total_profit DESC
LIMIT 10;
""", conn)
```

```python
pd.read_sql_query("""
SELECT
    customer_name,
    ROUND(AVG(total_sales), 2) AS avg_order_value
FROM sales
GROUP BY customer_name
ORDER BY avg_order_value DESC
LIMIT 10;
""", conn)
```

```python
pd.read_sql_query("""
SELECT
    segment,
    ROUND(SUM(total_sales), 2) AS total_revenue,
    ROUND(SUM(profit), 2) AS total_profit,
    COUNT(DISTINCT order_id) AS total_orders
FROM sales
GROUP BY segment
ORDER BY total_revenue DESC;
""", conn)
```

```python
pd.read_sql_query("""
SELECT
    segment,
    ROUND(SUM(profit), 2) AS total_profit,
    ROUND(SUM(total_sales), 2) AS total_revenue,
    ROUND((SUM(profit) / SUM(total_sales)) * 100, 2) AS profit_margin_percentage
FROM sales
GROUP BY segment
ORDER BY profit_margin_percentage DESC;
""", conn)
```

```python
pd.read_sql_query("""
SELECT
    customer_name,
    ROUND(SUM(total_sales), 2) AS total_revenue,
    ROUND(SUM(profit), 2) AS total_profit
FROM sales
GROUP BY customer_name
HAVING SUM(profit) < 0
ORDER BY total_profit ASC
LIMIT 10;
""", conn)
```
